```python
#NOW, PANDAS DOES NOT KNOW THAT DATE COLUMN HAS ITS OWN IMPORTANT MEANING HERE, SO I HAVE TO FIRST CONVERT THIS ATTRIBUTE IN
# ITS FORM
```

```python
avocados_df_copy['Date'] = pd.to_datetime(avocados_df_copy.Date)
```

```python
avocados_df_copy['year'] = pd.DatetimeIndex(avocados_df_copy.Date).year
avocados_df_copy['month'] = pd.DatetimeIndex(avocados_df_copy.Date).month
```
